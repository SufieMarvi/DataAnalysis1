```python
import matplotlib.pyplot as plt

# Create a boxplot of the Age column
plt.figure(figsize=(8, 6))
plt.boxplot(df['Age'].dropna())
plt.title('Boxplot of Age')
plt.show()

# Calculate Q1 (25th percentile), Q3 (75th percentile), and IQR
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

# Calculate the lower bound (anything below this is considered an outlier)
lower_bound = Q1 - 1.5 * IQR

# Count the number of outliers below Q1
outliers_below_Q1 = df[df['Age'] < lower_bound]

print(f'Number of outliers below Q1: {len(outliers_below_Q1)}')
```

```python
# prompt: Which column has a negative correlation with "Age"?

df_numeric = df.select_dtypes(include=np.number)
correlation_with_age = df_numeric.corr()['Age']
negative_correlation_columns = correlation_with_age[correlation_with_age < 0]

print("Columns with a negative correlation with Age:")
print(negative_correlation_columns)
```

```python
#Which column has the highest correlation with "Age"?

df_numeric = df.select_dtypes(include=np.number)
correlation_with_age = df_numeric.corr()['Age']
highest_correlation_column = correlation_with_age.idxmax()

print("Column with the highest correlation with 'Age':", correlation_with_age)
```
